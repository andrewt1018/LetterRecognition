```python
from emnist import extract_training_samples
```

```python
images, pre_labels = extract_training_samples('letters')
```

```python
pre_labels = list(pre_labels)
images = list(images)
```

```python
# Vectorizing labels
labels = []
for label in pre_labels:
    new_label = np.zeros(26)
    new_label[label - 1] = 1
    labels.append(new_label)
labels = list(labels)
```

```python
import numpy as np
import random
from matplotlib import pyplot as plt

dataset = list(zip(images, labels))
random.shuffle(dataset)
images, labels = (zip(*dataset))
images = np.asarray(images)
labels = np.asarray(labels)

rows = 5
cols = 2

fig = plt.figure(figsize=(10,7))
plt.rcParams["figure.figsize"] = [7.50, 3.50]
plt.rcParams["figure.autolayout"] = True

alphabet = list(chr(v) for v in range(ord('A'), ord('A') + 26))
for i in range(rows * cols):
    label = alphabet[labels[i].argmax()]
    
    fig.add_subplot(rows, cols, i + 1)
    plt.gray()
    plt.imshow(images[i])
    plt.axis('off')
    plt.title("Image of a {}".format(label))

plt.show()
```
